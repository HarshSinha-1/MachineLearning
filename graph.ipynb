```python
plt.plot(batsman["index"], batsman['V Kohli'])
plt.plot(batsman["index"], batsman['RG Sharma'])
plt.title("Rohit Sharma V/s Virat kohli Comparison of Runs")
plt.xlabel("ipl Seasons")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman["index"], batsman['V Kohli'], color = 'red')
plt.plot(batsman["index"], batsman['RG Sharma'], color = 'green')
plt.title("Rohit Sharma V/s Virat kohli Comparison of Runs")
plt.xlabel("ipl Seasons")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman["index"], batsman['V Kohli'], color = 'red',linestyle = 'dashed',linewidth = 3 )
plt.plot(batsman["index"], batsman['RG Sharma'], color = 'green',linestyle = 'dotted',linewidth = 4)
plt.title("Rohit Sharma V/s Virat kohli Comparison of Runs")
plt.xlabel("ipl Seasons")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman["index"], batsman['V Kohli'], color = 'red',linestyle = 'dashed',linewidth = 3,marker= 'D' )
plt.plot(batsman["index"], batsman['RG Sharma'], color = 'green',linestyle = 'dotted',linewidth = 4,marker= '>')
plt.title("Rohit Sharma V/s Virat kohli Comparison of Runs")
plt.xlabel("ipl Seasons")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman["index"], batsman['V Kohli'], color = 'red',linestyle = 'dashed',linewidth = 3,marker= 'D',label = 'virat' )
plt.plot(batsman["index"], batsman['RG Sharma'], color = 'green',linestyle = 'dotted',linewidth = 4,marker= '>', label ='rohit')
plt.title("Rohit Sharma V/s Virat kohli Comparison of Runs")
plt.xlabel("ipl Seasons")
plt.ylabel("Runs Scored")
plt.legend(loc = 'upper left')
```
