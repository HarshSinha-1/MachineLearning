```python
!pip install scikit-learn
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neighbors import LocalOutlierFactor
from sklearn.model_selection import train_test_split

# Load the dataset
data = pd.read_csv("/content/sample_data/california_housing_test.csv")

# 1. Dimensionality Reduction using PCA
# --------------------------------------
# Select numerical features for PCA
numerical_features = ['housing_median_age', 'total_rooms', 'total_bedrooms',
                     'population', 'households', 'median_income', 'longitude', 'latitude']

# Standardize the numerical features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data[numerical_features])

# Apply PCA to reduce dimensions (e.g., to 2 components)
pca = PCA(n_components=2)  # Choose the desired number of components
principal_components = pca.fit_transform(scaled_data)

# Create a DataFrame with the principal components
pca_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'], index=data.index)

# Concatenate with 'housing_median_age' if it exists
if 'housing_median_age' in data.columns:
    pca_df = pd.concat([pca_df, data[['housing_median_age']]], axis=1)

# 2. Instance Reduction using Local Outlier Factor (LOF)
# -----------------------------------------------------
# Apply LOF to identify outliers
lof = LocalOutlierFactor(contamination=0.1)  # Adjust contamination as needed
outlier_predictions = lof.fit_predict(scaled_data)

# Filter out the outliers (keep only inliers)
inliers_mask = outlier_predictions != -1  # -1 indicates outliers
reduced_data = data[inliers_mask]

# Combine dimensionality and instance reduction
# --------------------------------------------
final_reduced_data = reduced_data.copy()
final_reduced_data[['PC1', 'PC2']] = pca_df.loc[inliers_mask, ['PC1', 'PC2']]

# Now, final_reduced_data contains the data with reduced dimensions and instances.
```
