```python
dir_col = data['Debt.To.Income.Ratio']
```

```python
def custom_dir(dir_col):
      temp = dir_col.str.replace('%','')
      num = pd.to_numeric(temp)

      return num
custom_dir(dir_col)
```

```python
# a = data['FICO.Range'].str.split('-',expand=True)
# print(a)

b = data['FICO.Range']
```

```python
def custom_dir2(dir_col):
      temp = dir_col.str.split('-',expand=True) #expand
      num = (pd.to_numeric(temp[0], errors='coerce')+pd.to_numeric(temp[1], errors='coerce'))/2
      return num
custom_dir2(b)
```

```python
def custom_el(el_col):
  temp = el_col.replace({'5 years': 5,'4 years':4,'< 1 year':0,'10+ years':10,'1 year':1,'3 years':3,'9 years':9,'7 years':7})
  num = pd.to_numeric(temp,errors='coerce')
  return num

custom_el(a)
```

```python
cat_to_num_cols = ['Amount.Requested','Amount.Funded.By.Investors','Open.CREDIT.Lines','Revolving.CREDIT.Balance']
# columns which contain numeric info but only require simple typecasting
cat_to_dummies_cols=['Loan.Length','Loan.Purpose','State','Home.Ownership']
# groups of columns for which we need to create dummies
simple_num_cols=['Monthly.Income','Inquiries.in.the.Last.6.Months']
# columns which are all ready numeric and contain numeric information
custom_func_dict_cols={'Debt.To.Income.Ratio':custom_dir,'FICO.Range':custom_dir2,'Employment.Length':custom_el}
# columnswhich require custom processing to extract numeric information from the data
```

```python
ld_pipe = mt.DataPipe(cat_to_dummies=cat_to_dummies_cols,
                  cat_to_num=cat_to_num_cols,
                  simple_num=simple_num_cols,
                  custom_func_dict=custom_func_dict_cols)
```

```python
ld_pipe.fit(data)
```

```python
x_train=ld_pipe.transform(data)
```
